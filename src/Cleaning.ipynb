### Importando as bibliotecas

```python
from pandas import read_excel
```

### Leitura dos Dados

```python
df = read_excel("../Data/Sintomas Musculoesqueléticos em Universitários.xlsx")
df['IDADE'] = df['IDADE'].astype(str)
```

### Primeiras impressões dos dados

```python
df.head(5)
```

### Removendo todas as linhas onde os valores de todas as colunas são NaN

```python
len(df)



df = df[df[[str(x) for x in df.columns]].notnull().all(1)]  #Filtro que seleciona dados não nulos


df.drop(
    columns=['NOME COMPLETO',
    'Carimbo de data/hora',
    'CONCORDOU COM OS TERMOS'], inplace = True
)
```

### Conhecendo melhor valores possíveis em cada coluna para elaboração da análise de dados

```python
map_col_values = lambda df: {col: df[col].unique().tolist() for col in df.columns}
```

```python
df.columns
```

```python
valores_unicos = map_col_values(df)
```

```python

for k,v in valores_unicos.items():
    print(f'{k} ==> {v} \n\n')

```

### Removendo cópias de valores categóricos escritas de formas diferentes

```python
# Funções de normalização:

farmacia_norm = lambda curso : 'Farmácia' if curso in set(['Farma ', 'Farmácia', 'Farmacia ', 'Farmácia ']) else curso

direito_norm = lambda curso : 'Direito' if curso in set([ 'Direito ', 'direito ', 'Direito', 'Direito ' 'direito ', 'direito']) else farmacia_norm(curso)

eng_norm = lambda curso : 'Engenharia da Computação' if curso in set(['Engenharia da computação ', 'Engenharia de computação']) else direito_norm(curso)

medicina_norm = lambda curso : 'Medicina' if curso in set(['Medicina', 'Medicina ', 'medicina', 'medicina ']) else eng_norm(curso)

psicologia_norm = lambda curso : 'Psicologia' if curso in set(['Psicologia ', 'PSICOLOGIA ', 'psicologia ', 'Psicologia - Formação de Psicologo', 'Psicologia', 'psicologia', 'piscologia']) else medicina_norm(curso)

df['CURSO'] = df['CURSO'].map(psicologia_norm) #Filtrando dados repetidos

somente_numeros_da_idade = lambda idade : idade.join(s.strip() for s in idade.split())[:2]

df['IDADE'] = df['IDADE'].map(somente_numeros_da_idade)

```

```python
#Verificando filtragem
df['CURSO'].unique().tolist()
df['IDADE'].unique().tolist()
```
